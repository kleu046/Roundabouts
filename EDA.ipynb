```python
import pandas as pd

df = pd.read_csv('roundabouts_clean.csv')
```

```python
df.describe()
```

```python
categoricals = df.dtypes == 'object'
df.loc[:, categoricals].describe()
```

```python
akld = df[df.state_region.fillna('').str.contains('Auckland')]
akld.describe()
```

```python
df.loc[df.name.str.contains('Place Charles de Gaulle'),'control_type'] = 'All-Way Yield'
df.loc[df.name.str.contains('Place de la Nation'), 'control_type'] = 'All-Way Yield'
df.loc[df.name.str.contains('Place de la Nation'), 'lane_type'] = '3+ Lanes'
df.loc[df.name.str.contains('Metropolitan Plaza'), 'name'] = 'South Alandele Circle'
df.loc[df.name.str.contains('Dupont Circle'), 'name'] = 'Dupont Circle'
```

```python
import pickle

with open('akld.pickle', 'wb') as f:
    pickle.dump(akld[akld.approaches >= 5], f)
```

```python
world = df[df.approaches >= 5].sort_values('approaches', ascending=False).iloc[[0,2,3,4,7,8],:]

with open('world.pickle', 'wb') as f:
    pickle.dump(world, f)
```

```python
nz =  df[df.country.str.contains('New Zealand') & (df.approaches > 5)]

with open('nz.pickle','wb') as f:
    pickle.dump(nz, f)
```
